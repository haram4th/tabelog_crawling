```python
# 타베로그 사이트 접속
main_url = 'https://tabelog.com/'
driver = wd.Chrome(executable_path='./chromedriver.exe')
driver.get(main_url)
shopsinfo = [] # 가게 상세 페이지 HTML을 저장하는 리스트
```

```python
# 타베로그에서 가게 리스트에 있는 곳을 검색해 세부정보 추출
info_title_list = []
info_contents_list = []
for area, shop in tqdm(zip(shoplist['dohukenjpn'], shoplist['jshopname'])) :
    driver.find_element_by_id("sa").clear()
    driver.find_element_by_id("sa").send_keys(area)
    time.sleep(2)
    driver.find_element_by_id("sk").clear()
    driver.find_element_by_id("sk").send_keys(shop)
    time.sleep(2)    
    driver.find_element_by_id("js-global-search-btn").click()
    time.sleep(3)

    box = driver.find_element_by_css_selector('.js-rstlist-info.rstlist-info')

    linkurl = box.find_element_by_css_selector('a').get_attribute('href') 
    print(linkurl)
    # 검색된 페이지에서 가게명, 주력음식, 평점, 리뷰수, 휴무정보 등 추출)
    # 찾은 가게 페이지로 이동
    driver.get(linkurl)
    time.sleep(5)

    # 만약을 대비해서 shopsinfo에 페이지 소스를 저장해 둠
    soup = bs(driver.page_source, 'html.parser')
    shopsinfo.append(soup)

    #가게 정보가 있는 테이블 부분만 가져와 정보 추출
    shopinfo = driver.find_element_by_id('rst-data-head')
    shopinfotable = shopinfo.find_elements_by_css_selector('table.c-table.c-table--form.rstinfo-table__table')
    info = ''
    title = []
    contents = []
    for i in shopinfotable:
        info = bs(i.get_attribute('innerHTML'), 'html.parser')
        title.append(info.find_all('th'))
        contents.append(info.find_all('td'))

    info_title = []
    info_contents = []
    for ttemp, ctemp in zip(title, contents):
        for t, c in zip(ttemp, ctemp):
            title = t.text
            title = title.strip().replace('\n', "").replace('\t',"")
            contents = c.text
            contents = contents.strip().replace('\n', "").replace('\t',"")
            info_title.append(title)
            info_contents.append(contents)
    info_title_list.append(info_title)
    info_contents_list.append(info_contents)
    
    driver.get(main_url)
```

```python
# 종료
driver.close()
driver.quit()
# import sys
# sys.exit()
```

## 가게 정보 HTML 소스를 pickle로 저장

```python
print(len(info_title_list))
print(len(info_contents_list))
```

```python
import pickle
with open('info_title_list.pickle', 'wb') as file:
    pickle.dump(info_title_list,file)
with open('info_contents_list.pickle', 'wb') as file:
    pickle.dump(info_contents_list,file)
```

```python
# 추출된 가게 정보를 딕셔너리로 만든 후 데이터 프레임으로 합침
shop_info_list = []
for title, contents in zip(info_title_list, info_contents_list):
    info_data = dict(zip(title, contents))
    shop_info_list.append(info_data) 

```
